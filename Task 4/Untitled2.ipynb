```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
df = pd.read_csv("/content/cleaned_sales_dataset.csv")
print(df.head())
df.info()
```

```python


analysis_df = df[['Customer_Age', 'Total_Amount_Clean']].dropna()

corr_value, p_value = pearsonr(
    analysis_df['Customer_Age'],
    analysis_df['Total_Amount_Clean']
)

print(f"Correlation Coefficient: {corr_value:.3f}")
print(f"P-value: {p_value:.5f}")
```
